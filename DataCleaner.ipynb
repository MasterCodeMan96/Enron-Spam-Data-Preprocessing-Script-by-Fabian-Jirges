# üìÑ Datenvorbereitung: Enron E-Mails f√ºr Spam-Erkennung

## ‚úÖ Ziel

Ziel dieses Schrittes war es, die urspr√ºnglichen E-Mails aus dem **Enron Spam Dataset** in ein strukturiertes, bereinigtes Format zu bringen, das f√ºr Machine Learning nutzbar ist ‚Äì insbesondere f√ºr die Klassifikation von Spam und Ham (nicht-Spam).

---

## üîç Vorgehen

1. **Einlesen von E-Mails aus Textdateien:**
   - Die Daten lagen im Original als `.txt`-Dateien in zwei Ordnern vor: `ham` (normale E-Mails) und `spam` (unerw√ºnschte E-Mails).
   - Jede E-Mail wurde als einzelne Datei gespeichert.

2. **Bereinigung der Inhalte:**
   - Entfernen von Betreffzeilen (`Subject:`).
   - Entfernen von Headern (Metadaten am Anfang der E-Mail).
   - Entfernen von nicht sichtbaren Steuerzeichen (z.‚ÄØB. `\x01`, `\x0b`), um Kompatibilit√§t mit Excel sicherzustellen.
   - Vereinheitlichung von Leerzeichen und Zeilenumbr√ºchen.

3. **Strukturierung in einem DataFrame:**
   - Alle E-Mails wurden zusammengef√ºhrt und in einem `pandas.DataFrame` gespeichert.
   - Jede Zeile enth√§lt den bereinigten Text einer E-Mail und das zugeh√∂rige Label (`spam` oder `ham`).

4. **Label-Encoding:**
   - Zus√§tzlich wurde eine numerische Spalte `label_num` erstellt:
     - `1` f√ºr Spam
     - `0` f√ºr Ham

5. **Export als Excel-Datei:**
   - Der finale Datensatz wurde als `.xlsx`-Datei gespeichert.
   - Pfad: `C:\Users\...\enron_bereinigt.xlsx`
   - Die Datei kann direkt f√ºr Training, Analyse oder Visualisierung verwendet werden.

---

## üìÅ Ergebnis

Die Excel-Datei enth√§lt drei Spalten:

| text (bereinigt) | label (Text) | label_num (numerisch) |
|------------------|--------------|------------------------|
| "Hallo, ich hoffe..." | ham |                |0 |
| "Sie haben gewonnen!" | spam |               |1 |

---

## üí° Warum dieser Schritt wichtig ist

- Machine-Learning-Modelle (z.‚ÄØB. Naive Bayes, BERT) ben√∂tigen **strukturierte und saubere Daten**.
- E-Mail-Daten enthalten h√§ufig **Rauschen**, Header-Informationen und Sonderzeichen, die das Modell negativ beeinflussen k√∂nnen.
- Durch diese Vorverarbeitung schaffen wir die Grundlage f√ºr **verl√§ssliches und reproduzierbares Modelltraining**.



```python
import os
import pandas as pd
import re

def clean_email(text):
    import string
    import re

    # Entferne Betreffzeilen
    text = re.sub(r'^Subject:.*\n?', '', text, flags=re.IGNORECASE | re.MULTILINE)

    # Entferne Header
    parts = text.split('\n\n', 1)
    text = parts[1] if len(parts) > 1 else parts[0]

    # Entferne Steuerzeichen (nicht druckbare ASCII-Zeichen)
    text = ''.join(ch for ch in text if ch in string.printable)

    # Entferne mehrfaches Leerzeichen
    text = re.sub(r'\s+', ' ', text).strip()

    return text


def read_and_clean_emails(folder_path, label):
    data = []
    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)
        if os.path.isfile(file_path) and filename.endswith('.txt'):
            with open(file_path, 'r', encoding='latin-1') as file:
                raw_text = file.read()
                clean_text = clean_email(raw_text)
                if clean_text:
                    data.append({'text': clean_text, 'label': label})
    return data

# Windows-Pfade mit raw strings!
ham_folder = r'yourpath'
spam_folder = r'yourpath'

# Einlesen & bereinigen
ham_emails = read_and_clean_emails(ham_folder, 'ham')
spam_emails = read_and_clean_emails(spam_folder, 'spam')

# Kombinieren in DataFrame
df = pd.DataFrame(ham_emails + spam_emails)

# Numerisches Label hinzuf√ºgen
df['label_num'] = df['label'].map({'ham': 0, 'spam': 1})

# Ausgabe pr√ºfen
print(df.head())

# Als Excel speichern
excel_path = r'yoursavepath'
df.to_excel(excel_path, index=False)

print(f"‚úÖ Excel-Datei erfolgreich gespeichert unter: {excel_path}")
```
